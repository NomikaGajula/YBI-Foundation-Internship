```python
df=pd.read_csv('https://raw.githubusercontent.com/ybifoundation/Dataset/main/Fish.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
yy=df[['Weight']]
```

```python
yy.shape
```

```python
xx=df.iloc[:,3:]
```

```python
xx.columns
```

```python
xx
```

```python
import statsmodels.api as sm
```

```python
xx=sm.add_constant(xx)
```

```python
xx.head()
```

```python
xx_train,xx_test,yy_train,yy_test=train_test_split(xx,yy)
```

```python
xx_train.shape,xx_test.shape
```

```python
Mod=sm.OLS(yy_train,xx_train).fit()
#here should not write as sm.OLS(xx_train,yy_train) error will arise as dimensions are not valid
```

```python
y_preddict=Mod.predict(xx_test)
```

```python
y_preddict.shape
```

```python
r2_score(yy_test,y_preddict)
```

```python
mean_absolute_percentage_error(yy_test,y_preddict)
```
