```python
import numpy as np
```

```python
a=[3,4,5,3,7,1]
```

```python
#to find the no of dimensions of an array a
print(np.ndim(a))
#to find no of elements in array a
print(np.size(a))
#to sort the elements in array
print(np.sort(a))
print(a)
```

```python
#to find maximum element in array a
print(np.max(a))
#to find minimum element in array a
print(np.min(a))
#to find the mean of elements in the array a
print(np.mean(a))
```

```python
#to get an array of n elements let n be 10
A=np.arange(0,10)
print(A)
#to get an array of n elements with a gap of d in between them let n be 10 and d be 2
B=np.arange(0,10,2)
print(B)
#to get d no of elements from n numbers
D=np.linspace(0,10,3)
print(D)
#to get 5 random numbers in between range 0 to n
E=np.random.randint(0,20,5)
print(E)
```

```python
#to get a 2 dimentional array
C=np.arange(0,10).reshape(2,5)
print(C)
C.ndim
```

```python
#to find sin,sqrt,cos of array a elements
print(np.sin(A))
print(np.cos(A))
print(np.sqrt(A))
```

```python
#to convert the values in array a to degrees
np.degrees(A)
```

Matplotlib

```python
import matplotlib.pyplot as plt
```

This is one way of getting pie plot along with the label names

```python
data.sqft.plot(kind='pie')
```

```python
data.sqft.plot.pie()
```

```python
#For data visualisation we use matplotlib
#to represent one of the column in data as piechart
a=[3,4,5,6,2]
plt.pie(a,[0,0,0.4,0,0],colors=['red','orange','pink','yellow','purple'],labels=[3,4,5,6,2])
plt.legend(loc='upper left') #legend represents the index of each part in piechart and what parameter does it represent
plt.show()
```

```python
plt.bar(data.floor,data.totalFloor,color='red')
plt.title('Bar Graph',color='blue',fontsize=14)
```

```python
#one way of getting pie plot with out manullay adding labels
data.place.value_counts().plot.pie()
plt.legend(loc='best')
```

```python
#getting horizontal barplot along with respective label names
data.place.value_counts().plot.barh()
```

```python
data.place.value_counts().plot(kind='bar')
```

```python
plt.plot(data.floor,data.totalFloor,linestyle='dashed',color='orange',linewidth=1.5,marker='P',markersize=7.5)
plt.xlabel('floor',color='blue',fontsize='14')
plt.title("Line plot with all various representations",color='blue',fontsize=16)
plt.ylabel('totalFloor',color='blue',fontsize=14)
```

```python
#correlation is the strength of relation of one parameter with another parameter
#annot represent the respective values of correlation in the each box 
sb.heatmap(data.corr(),annot=True)
```
